# [Cursor Tokens](https://pro.arcgis.com/en/pro-app/arcpy/data-access/searchcursor-class.htm)

Geometry properties can be accessed by specifying the token SHAPE@ in the list of fields.

```python
import arcpy
import os
```

```python
gdb = r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Week 5\Crime.gdb'
feature_class = "December2017_Crime"

fc = os.path.join(gdb, feature_class)
print(fc)
```

Let's print out the geomerry as:
- XY coordinates
- [JSON](https://www.json.org/json-en.html)
- [WKT](https://en.wikipedia.org/wiki/Well-known_text_representation_of_geometry)
- Area
- Length

```python
fields = ['SHAPE@XY', 'SHAPE@AREA', 'SHAPE@LENGTH', 'SHAPE@JSON', 'SHAPE@WKT']

# For each row print the WELL_ID and WELL_TYPE fields, and the
# the feature's x,y coordinates
with arcpy.da.SearchCursor(fc, fields) as cursor:
    for row in cursor:
        #print row[0]
        if row[2] != None:
            print('SHAPE@XY: ' + str(row[0]))
            print('SHAPE@AREA: ' + str(row[1]))
            print('SHAPE@LENGTH: ' + str(row[2]))
            print('SHAPE@JSON: ' + str(row[3]))
            print('SHAPE@WKT: ' + str(row[4]))
```

# Creating Geometries and Inserting them Into a New Feature Class

```python
import arcpy
import os
```

```python
fc =r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Week 5\Polygon.gdb\features'
gdb_path = os.path.dirname(fc)
fc_name = os.path.basename(fc)
gdb_folder = os.path.dirname(gdb_path)
gdb_name = os.path.basename(gdb_path)
print(gdb_name)
print(gdb_folder)
```

## Create a new file geodatabase

```python
arcpy.CreateFileGDB_management(gdb_folder, gdb_name)
```

## Create a new feature class

```python
arcpy.CreateFeatureclass_management(gdb_path, fc_name,
                                "POLYGON", "#", "DISABLED",
                                "DISABLED", arcpy.SpatialReference(3857))
```

```python
print("Insert Cursor")
fields = ('SHAPE@')
cursor = arcpy.da.InsertCursor(fc, fields)

rectangle = [(0, 533.3),
         (1000, 533.3),
         (1000, 0),
         (0, 0)]
cursor.insertRow([rectangle])

del cursor
print("Done.")
```
