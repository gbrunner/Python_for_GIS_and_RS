````python
import arcpy

arcpy.env.workspace = r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Projects\imagery'

raster_list = arcpy.ListRasters()
print("There are " + str(len(raster_list)) + " rasters in the folder.")
````

````python
for raster in raster_list:
    desc = arcpy.Describe(raster)
    print("Raster Name: " + raster)
    print("Band Count:       %d" % desc.bandCount)
    print("Compression Type: %s" % desc.compressionType)
    print("Raster Format:    %s" % desc.format)
    print("Spatial Reference: %s" % desc.SpatialReference.name)
````

# Problem 2 (15 Points)

Unzip **tlgdb_2019_a_us_school.gdb.** Create a Notebook that uses arcpy to create a report about the features in the geodatabase. List out the following:
- Feature class name
- Number of features in the feature class
- Projection of the feature class
- Shape type of the feature class (point, polyline, polygon, etc.)

**Bonus (5 points)** Write your results out to a text file or comma separated text file.

````python
import arcpy

arcpy.env.workspace = r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Projects\tlgdb_2019_a_us_school.gdb'

feature_list = arcpy.ListFeatureClasses()

f = open('results.txt', 'w')
header = f.write('Feature Class, Feature Count, Projection, Shape')

for feature in feature_list:
    feature_count = arcpy.GetCount_management(feature)
    desc = arcpy.Describe(feature)
    projection = desc.SpatialReference.name
    shape_type = desc.shapeType
    seperator = ", "
    print(feature + seperator, str(feature_count) + seperator, projection + seperator, shape_type)
    f.write(feature + seperator + str(feature_count) + seperator + projection + seperator + shape_type + '\n')

f.close()
````

### Looping over multiple queries

````python
import os
import arcpy
import random

fc = r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Projects\usa_cities.gdb\largest_cities'
workspace = r'C:\Users\greg6750\Documents\IPython Notebooks\intro-prog-for-gis-rs\Projects'
gdbname = 'problem5.gdb'
arcpy.env.workspace = os.path.join(workspace, gdbname)

# Set overwrite option
arcpy.env.overwriteOutput = True

pop_query = ["POPULATION > 1000000",
             "POPULATION >= 1000000 AND POPULATION <= 500000" , 
             "POPULATION >= 500000 AND POPULATION <= 100000",
             "POPULATION >= 100000 AND POPULATION <= 10000"]

for query in pop_query:
    print("Processing query: " + query)
    min_pop_break = query.split(' ')[-1]
    arcpy.MakeFeatureLayer_management(fc,"cities_lyr_" + str(min_pop_break))

    arcpy.SelectLayerByAttribute_management ("cities_lyr_" + str(min_pop_break), "NEW_SELECTION", query)
        
    arcpy.CopyFeatures_management("cities_lyr_" + str(min_pop_break), 
                                  "pop__greater_than_" + str(min_pop_break))
print("Done")
````
