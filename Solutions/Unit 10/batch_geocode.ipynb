## Import batch_geocode

```python
from arcgis import GIS
gis = GIS("http://slustl.maps.arcgis.com/home", "gregbrunner_slugis")
```

```python
from arcgis.geocoding import batch_geocode
```

```python
import pandas as pd
```

```python
## Read CSV to data frame
```

```python
df = pd.read_csv('addresses.csv')
df.head()
```

## Import as Arcgis content

```python
addresses = gis.content.import_data(df, {"Address" : "LOCATION"})
```

## Geocode the addresses

```python
results  = batch_geocode(df['Single Line Address'])
```

```python
results
```

## Create a new map

```python
my_map = gis.map()
my_map
```

## Draw the x,y coordinates to the map

```python
for address in results:
    my_map.draw(address['location'])
```

## Save the map

```python
webmap_properties = {'title':'Module 10 - Exercieses GJB',
                    'snippet': 'Geocoding and ArcGIS Python API Practice',
                    'tags':['GIS', 'python', 'practice']}
saved_webmap = my_map.save(webmap_properties)
```
